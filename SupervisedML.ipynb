# Task- 2 Prediction using Supervised ML

**Importing Libraries**

```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/AdiPersonalWorks/Random/master/student_scores%20-%20student_scores.csv")
df.head()
```

```python
df.info()
```


**Exploratory Data Analysis(EDA)**

```python
sns.displot(df.Hours,color="r")
```

```python
sns.displot(df.Scores,color="b")
```



**Relationship between the Data**

```python
# Plotting the distribution of scores
sns.jointplot(x="Hours",y="Scores",data=df,kind="scatter",color="c", height=6)
```

**Graph shows a positive linear relation between the number of hours studied and percentage of score.**




# X-Y Split

```python
X=df[["Hours"]]
print(type(X))
print(X.shape)
X.head()
```

```python
Y=df.Scores
print(type(Y))
print(Y.shape)
Y.head()
```

# Test-Train Split

```python
from sklearn.model_selection import train_test_split 
X_train, X_test, Y_train, Y_test = train_test_split(X, Y,test_size=0.2, random_state=0)
```

# Training Linear Regression

```python
from sklearn.linear_model import LinearRegression  
linreg = LinearRegression()  
linreg.fit(X_train, Y_train)
```



**Interpreting Model Coefficient**

```python
print(linreg.intercept_)
print(linreg.coef_)
```

```python
# Plotting the regression line
line = linreg.coef_*X+linreg.intercept_

# Plotting for the test data
plt.scatter(X, Y,color="m")
plt.plot(X, line, color="c");
plt.show()
```

# Making Predictions with the Model

```python
Y_pred = linreg.predict(X_test)
Y_pred[:5]
```

```python
# Comparing Actual vs Predicted
df1 = pd.DataFrame({'Actual': Y_test, 'Predicted': Y_pred})  
df1
```

```python
sns.jointplot(x="Actual",y="Predicted",xlim=(0,100),ylim=(0,80),data=df1,kind="reg",color="hotpink")
```

**The Plot shows the relationship between Actual and predicted scores.**



# Problem Statement:
**What will be predicted score if a student study for 9.25 hrs in a day?**

**Answer:**

```python
hours = 9.25
my_pred = linreg.predict([[hours]])
print("Predicted Score of a student who studies for {} hours a day is {} ".format(hours,my_pred[0]))
```



**Predict on any value**

```python
#input from user
hours = float(input("Enter the number of hours a student studied:"))
my_pred = linreg.predict([[hours]])
print("Predicted Score of a student who studies for {} hours a day is {} ".format(hours,my_pred[0]))
```


# Model Evaluation Matrics

```python
from sklearn import metrics  
print('Mean Absolute Error:', metrics.mean_absolute_error(Y_test, Y_pred))
```

```python
print('Mean Squared Error:', metrics.mean_squared_error(Y_test, Y_pred))
```

```python
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(Y_test, Y_pred)))
```
